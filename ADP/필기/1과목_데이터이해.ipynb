## 1과목 데이터의 이해

### 데이터 유형
- 정성적 데이터(Qualitative data): 범주형$\cdot$비정형 데이터, 질적 척도
- 정량적 데이터(Quantitative data): 수치형$\cdot$정형 데이터, 양적 척도

### 데이터셋 유형
- 정형 데이터: 형태가 있는 데이터
  - 관계형 DB에 저장된다.
  - 예시)ERP, CRM, SCM, 정보시스템 등
- 반정형 데이터: 형태가 스키마, 메타데이터로 있는 데이터
  - 파일(XML, HTML, JSON 등)로 저장된다.
  - 데이터 분석 시 메타데이터가 있어야 분석결과를 해석할 수 있다.
  - 예시)모바일 데이터, 센싱 데이터 등
- 비정형 데이터: 형태가 없는 데이터
  - NoSQL, 객체지향DB에 저장된다.
  - 특정 처리 과정을 거쳐야 분석이 가능하다.
  - 예시)영상, 음성, 문자, SNS, 멀티미디어, **로그** 데이터, 사용자정의데이터 등

### <font color="forestgreen">데이터 가치 구조(DIKW 피라미드)
**데이터**: 객관적인 사실<br>
$~~~~~~~~~~~$ 예시)A마트는 100원, B마트는 200원에 연필을 판매한다.<br>
$\rightarrow$ **정보(개념)**: 데이터 가공, 처리 및 상관`관계`간 이해를 통해 패턴을 인식하고 의미를 부여한 것<br>
$~~~~~~~~~~~$ 예시)A마트의 연필이 더 싸다.<br>
$\rightarrow$ **지식**: 다양한 정보 중 유의미한 정보간 패턴을 이해하고 예측한 결과물(행동)<br>
$~~~~~~~~~~~$ 예시)상대적으로 저렴한 A마트에서 연필을 사야겠다.<br>
$\rightarrow$ **지혜**: `개인`의 지식, 경험 등이 축적되고 구조화된 창의적 아이디어<br>
$~~~~~~~~~~~$ 예시)A마트의 다른 상품들도 B마트보다 쌀 것이라 판단한다.

### <font color="forestgreen">데이터의 구축 체계
- 데이터 레이크: 발생하는 모든 데이터가 저장되는 시스템
- 데이터 베이스: Source에 무관하게 수집$\cdot$축적된 여러 관련 데이터를 일관된 체계로 구축한 시스템
  - 동시에, 복수의 이용자들이 사용할 수 있도록 편성되어 있다.
  - 중복 데이터가 없고 현재의 정확한 데이터 유지 하에 새 데이터 삽입$\cdot$기존 데이터 삭제$\cdot$갱신이 이루어져 DB의 데이터는 통합$\cdot$변화되는 데이터이다.
  - 기업 내부 데이터베이스는 아래와 같이 변화하였다.<br>
    OLAP: 다양한 비지니스 관점에서 다차원적인 데이터에 접근하여 의사 결정에 활용할 수 있는 정보를 얻을 수 있게 하는 기술<br>
    $~~~~~~~~~~$ 예시)판매 추이, 구매 성향, 재무 회계분석<br>
$\rightarrow$ ERP: 생산성을 극대화하기 위해 독립적으로 운영되던 경영자원을 하나로 재구축한 통합 시스템 <br>
$~~~~$ CRM: 고객의 특성$\cdot$가치 분석을 기반으로 한 고객 중심 자원 극대화$\cdot$고객 중심 마케팅까지의 데이터를 구축한 시스템<br>
$~~~~$ SCM: 공급망 단계 최적화의 데이터 시스템<br>
$~~~~$ BI: 기업 데이터를 분석해 기업의 의사결정에 활용하는 일련의 과정

$~~~$ ※ DBMS: 데이터 베이스 관리(데이터 검색$\cdot$저장$\cdot$구축), 프로그램과 데이터 베이스 간 공유$\cdot$사용 환경 제공을 하는 소프트웨어
- 데이터웨어하우스: 여러 데이터 베이스에서 `주제 중심적`으로 특정 조직의 특정 업무와 관련한 데이터만을 추출$\cdot$가공해 저장한 시스템
  - 주제중심성, 영속성$\cdot$비휘발성(읽기모드로 변경 불가), 통합성, 시계열성을 갖는다.
  - 일반 사용자들도 쉽게 데이터에 접근해서 원하는 분석을 수행할 수 있도록 간결, 정확하도록 테이블 모델링을 수행한다.
    - 스타 스키마 방법: 단일 사실 테이블(제3정규형)을 중심으로 다수의 차원 테이블(제2정규형)을 구성하는 방법
    - 스노우 플레이크 스키마 방법: 단일 사실 테이블(제3정규형)을 중심으로 다수의 차원 테이블(제3정규형)을 구성하는 방법
    - 차원테이블의 비정규화로 스타 스키마는 복잡도가 낮아 이해하기 쉽고 쿼리 작성 용이, 조인 테이블 개수가 적다.
    - 차원테이블의 정규화로 스노우 플레이크 스키마는 데이터 중복이 제거돼 스타 스키마에 비해 데이터 적재 시 적은 시간이 소요된다.
- 데이터 마트: 하나의 주제, 부서 중심의 데이터 웨어하우스


### 빅데이터
3V(양, 다양성, 수집$\cdot$처리 속도)를 갖춘 데이터
- 처리$\cdot$분석 기술, 인재$\cdot$조직 변화까지 아울러 정의된다.
- 다양한 비지니스에 활용됨에 공동 활용 목적으로 구축된 구조물인 플랫폼에 비유되며<br>
  산업 전반의 생산성을 끌여올려 산업혁명의 석탄$\cdot$철(혁명적 변화), 21세기의 원유(신규 산업 출현 비유)와 산업 발전에의 영향으로 렌즈로 비유된다.

### 빅데이터로 인한 변화
- 사후처리, 전수조사를 실시하고 질보다 양, 데이터간 상관관계를 중시하게 되었다.
- 사생활 침해 통제와 명확한 결과에 대해 책임을 묻는 책임 원칙 훼손, 잘못된 믿음, 적용 같은 데이터 오용 문제가 붉어져<br>
  개인정보 사용자가 피해 책임을 지게 함, 개인정보 비식별 기술, 알고리즘 부당함 반증 등의 방법이 강구되었다.<br>
  ※ 개인정보 비식별 기술 예시
  - 데이터 마스킹: 데이터의 길이, 유형, 형식 같은 속성은 유지한 채 중요 정보를 특수기호로 가리는 기술
  - 데이터값 삭제: 개인식별에 중요한 값, 필요없는 값을 삭제하거나 상위 단위로 표현하는 기술
  - 데이터값 대체$\cdot$변환: 가명처리, 총계처리, 범주화 등의 기술

### 빅데이터 활용
- 유전자 알고리즘: `최적화`가 필요한 문제의 해결책을 자연선택, 돌연변이 등과 같은 방식으로 점진적으로 찾아나가는 방법
- 유형분석: 특성, 기준을 찾아 분류, 그룹화하는 방법
- 소셜네트워크분석(사회관계망분석): 사람 간 관계, 영향력 있는 사람을 분석하는 방법
- 연관규칙탐사(윌마트 구매패턴분석), 기계학습, 회귀분석, 감정분석(선거 활동)
  
※ 우수고객, 이탈 예측, 구매패턴 분석 등 빅데이터가 필요하지 않은 경우도 있다. 빅데이터를 적재적소함에 비로소 전략적 통찰을 이끌 수 있다.

### 빅데이터 인재
- 이론$\cdot$분석 기술 같은 Hard Skill과 분석$\cdot$전달$\cdot$소통$\cdot$협력 같은 Soft Skill을 두루 갖춰야 한다.
- 단순세계화(컨버전스)에서 복잡세계화(디버전스)로의 변화, 생산에서 서비스$\cdot$시장창조(무형자산, 지식자산)로의 관점 변화에 의해<br>
  다양한 관점으로 바라보고 시장, 문제 이해$\cdot$파악하는 능력인 Soft Skill이 더 강조되며 인문학이 열풍이다.
- 기회 발굴, 강력한 성장동력을 가능하게 하는 전략도출을 위한 가치기반 분석을 할 수 있어야 한다.<br>
  ※ 일차원적 분석으로는 변화 대응$\cdot$파악, 새로운 기회 포착이 어렵다.<br>
