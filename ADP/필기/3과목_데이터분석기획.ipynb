### 분석 방법론
체화된 데이터분석 절차,방법(암묵지, 내면화$\cdot$공통화)을 형식화한 문서(형식지, 표출화$\cdot$연결화)를 체계화한 자료이다.
- 데이터분석을 기업 전사에 정착,체화시키기 위해 작성한다.<br>
  ※ 분석 방법론 정착의 장애요소로 **프레이밍 효과**(문제의 표현방식에 따라 동일한 사건$\cdot$상황이나 개인의 판단$\cdot$선택이 달라질 수 있는 현상) 등이 있다.
- 분석방법론 모델
  - 폭포수 모델: 단계를 반드시 순차적으로 진행하는 방법
  - 프로토타입 모델: 일부분을 우선 개발하고 사용자에게 제공한 뒤 요구$\cdot$성능 분석을 반복하며 점진적으로 개발하는 방법
    - 고객의 요구를 완전하게 이해하고 있지 못하거나 완벽한 요구 분석의 어려움을 해결한다.
  - 나선형 모델: 반복을 통해 점증적으로 개발하는 방법
    - 처음 시도하는 프로젝트에는 적용이 용이하나 관리 체계를 효과적으로 갖추지 못한 경우 복잡도가 상승해 프로젝트 진행이 어렵다.
데이터분석을 효과적으로 기업 내 정착시키기 위해 체계화한 절차와 방법이다.

### 분석기획
분석을 수행하기 전 분석 과제$\cdot$가치 정의, 분석 수행$\cdot$위험 관리 계획 수립, 결과 관리 방안 등을 계획하는 과정이다.
#### 1. 분석 주제 
자료로부터 분석대상(분석목표) 설정, 분석대상에 적용할 분석방법이 정해진 것의 여부로 4가지 분석주제 중 선택 등 분석의 방향$\cdot$범위를 설정하는 단계이다.

|  |  |  |
|:---:|:---:|:--:|
|        | 분석대상 Known | 분석대상 Unknown |
| 분석방법 Known | Optimization | **Insight** |
| 분석방법 Unknown | **Solution** | Discovery |

#### 2. 분석 과제 발굴
$\blacktriangleright$ 하향식 접근 방법: 문제가 주어진 상태에서 분석이 이루어지는 방법
- **문제 탐색**: 비지니스 모델과 외부 참조 모델로 분석 가치가 발생하는 모든 문제를 빠짐없이 도출하는 단계
  - 비지니스 모델을 활용해 전통적인 영역인 업무, 제품, 고객, 규제와 감사, 지원 인프라와 새로운 영역인 거시적인 메가트랜드(STEEP), 경쟁자 확대,<br> 시장니즈 탐색(운영채널, 고객에게 상품$\cdot$서비스를 전달하는 경로 등), 역량의 재해석에 대해 주제를 도출한다.
  - 외부 참조 모델을 통한 문제 탐색은 가치발굴 사례를 정리해둔 분석 테마 후보 그룹(Pool)에서 벤치마킹해 분석기회를 발굴하는 방법이다.
  - 도출한 분석기회들은 추후 분석 문제로 전환, 적합성 평가를 위해 문제, 가치 등을 명시하는 분석 유즈케이스를 작성해야 한다.
- **문제 정의**: 문제 탐색에서 도출된 비지니스 문제를 데이터 분석하기 위해 필요한 데이터와 기법을 정의하는 단계
- **해결방안 탐색**: 문제정의의 데이터, 기법을 확보하기 위해 분석 도구, 분석 교육 등의 해결방안을 모색하는 단계 
|  |  |  |
|:----------:|:-------:|:-------:|
|        | 분석역량 확보 | 분석역량 미확보 |
| 분석기법 및 시스템 기존 | 기존시스템 개선 활용 | 교육 및 채용을 통한 역량 확보 |
| 분석기법 및 시스템 신규 | 시스템 고도화 | 전문업체 Sourcing |


- **타당성 검토**: 경제적 타당성(비용 대비 편익), 데이터 및 기술적 타당성(분석실현)을 분석해 가장 우월한 문제를 택하는 단계

$\blacktriangleright$ 상향식 접근 방식: 데이터 분석(What)을 통해 문제가 발생한 상황을 역추적해 문제 도출, 재정의하는 방법
- 최근 복잡해지고 다양해진 환경 변화에 적합하지 않고 새로운 문제 탐색에 한계가 있는 하향식 접근 방식의 대응책이다.
- 문제, 목적이 정해지지 않은 비지도 학습 방법은 상향식 접근 방식에 해당한다. 
- 프로토타이핑 접근법은 완전하지는 못해도 신속하게 해결책을 제공하고 반복적으로 보완하는 방법으로 반복을 통해 문제 인식 수준을 높이고 필요데이터 존재의 불확실성에 대처(불가능한 프로젝트 수행 위험 방지), 데이터 사용 목적과 범위, 가치를 확대시킬 수 있다.
  
※ 디자인 사고: 상향식 접근방식의 발산과 하향식 접근방식의 수렴의 상호보완적인 의사결정 방법<br>
※ 분석과제를 발굴한 후 분석별로 필요한 데이터, 분석방법, 데이터 입수 및 분석 난이도, 분석 수행주기, 분석결과에 대한 검증 오너십, <br>$~~~$상세 분석과정 등을 정의한 분석과제정의서를 작성해야 한다.

#### 3. 분석 우선순위 설정
도출된 분석과제들을 전략적중요도, ROI(비지니스 성과), 실행 용이성로 평가해 우선순위를 정한다.
- 포트폴리오 사분면 분석
  - 시급성, 난이도로 평가한다.
    - 시급성: 현재와 미래의 전략적 가치 중 어느 것이 목표가치와 부합하는 지 정하는 요소, 분석 가치인 ROI의 비지니스 효과 요소
    - 난이도: 과제 추진에 필요한 비용, 분석범위가 진행이 용이한지, 준비가 필요한지를 평가하는 요소, <br>
      $~~~~~~~~~~~$ 빅데이터의 3V측면을 다루는 데 필요한 비용, 기술인 ROI의 투자비용 요소
      
     분석과제 적용 우선순위 기준을 시급성에 두면 Ⅲ$\rightarrow$Ⅳ$\rightarrow$Ⅱ, 난이도에 두면 Ⅲ$\rightarrow$Ⅰ$\rightarrow$Ⅱ순으로 의사결정 한다.
  - 분석 범위$\cdot$기술 변경으로 우선순위를 조정할 수 있다.
- 정보전략계획(ISP): 기업 및 공공기관에서 조직 내$\cdot$외 환경 분석, 문제점, 사용자 요구를 분석해 시스템의 중장기 로드맵을 수립하는 절차
    
|         | 당면한 분석 주제의 해결(과제 단위) | 지속적 분석 문화 내재화(중장기적 마스터플랜 단위)  |
|:-------:|:--:|:----------:|
| 1차목표 | Speed&Test | Accuracy&Deploy |
| 과제유형 | Quick&Win | Long Term View |
| 접근방식 | Problem Solving  | Problem Definition |

#### 4. 위험 관리 계획
분석 과정 내 위험을 식별하고 회피, 전이, 완화, 수용으로 위험 관리를 계획하는 단계이다.

### 분석 거버넌스
데이터 기반의 기업 의사결정 문화 정착, 분석업무 고도화를 **목적**으로 데이터 분석, **활용**을 체계적으로 관리하는 지속적인 프로세스다.
#### 1. 데이터 분석 수준진단
데이터 분석 기반을 구현하기 위해 준비, 보완할 부분을 찾고 분석 유형 및 방향성을 결정하기 위해 진단하는 단계이다.
- 분석 준비도: 분석업무파악, 인력 및 조직, 분석기법, 분석 데이터, 분석문화, IT인프라 영역의 수준을 진단해 분석도입 수준을 파악하는 방법
- 분석 성숙도: CMMI 모델을 사용해 비지니스, 조직$\cdot$역량, IT부문에 대해 수준진단 후 도입,활용,확산,최적화단계로 평가하는 방법
- 두 지표를 바탕으로 사분면 분석해 데이터 분석의 목표방향을 정의, 개선방안 수입한다.

  
#### 2. 데이터 거버넌스 체계 수립
전사의 모든 데이터 관리, 표준화, 운영조직 및 책임 등 표준화된 관리체계와 운영을 위한 프레임워크, 저장소를 구축하는 체계를 수립하는 단계이다.
- 원칙, 조직, 프로세스를 비지니스 목적에 맞게 유기적으로 조합해 관리한다.
- 빅데이터 거버넌스의 경우 다양한 데이터의 관리체계, 데이터 최적화, 데이터 생명주기 관리, 데이터 카테고리별 관리책임자 지정 등을 포함한다.
- **데이터 표준화**: 데이터 표준용어 설정, 명명규칙 수립(언어별 매핑상태 유지해야 함), 메타 데이터(MD), 데이터 사전(DD)을 구축한다.<br>
  $\rightarrow$ **데이터 관리체계**: 데이터 정합성, 활용 효용성을 위해 표준데이터를 포함한 MD. DD의 관리원칙과<br>
  $~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ 데이터 가용성 및 관리비용을 고려해 데이터 생명주기 관리방안을 수립한다.<br>
  $\rightarrow$ **데이터 저장소 관리**: MD 및 표준데이터를 관리하기 위한 저장소(워크플로우와 응용 소프트웨어 지원)를 구성하고 <br>
  $\qquad\qquad\qquad\quad~~$ 데이터 구조 변경에 따른 사전영향평가를 수행한다.<br>
  $\rightarrow$ **표준화 활동**: 거버넌스 체계의 표준 준수 여부를 주기적으로 점검, 개선, 교육해 실용성을 높인다.

#### 3. 데이터 조직 및 인력방안 수립
분석기회 발굴, 타부서간 협조$\cdot$지원, 분석 수행 등에 효율적인 조직구조와 인력구성을 택하는 단계이다.
- 조직구조
  - **집중구조**: 분석전담조직에서 분석을 수행하는 구조
    - 전략적 중요도를 논해 우선순위에 따라 분석을 진행할 수 있으나 현업 업무부서의 분석업무와 이중화/이원화 가능성이 높다.
  - **기능구조**: 분석조직없이 각 업무 부서 내에서 분석을 수행하는 구조
    - 전사 차원에서의 분석이 어려우며 과거실적에 국한된 분석 수행 가능성이 높다.
  - **분산구조**: 분석조직과 각 업무 부서 모두에 분석조직인력이 배치된 구조
    - 전사차원의 우선순위 분석, 분석결과 공유$\cdot$신속한 의사결정이 가능하다. 
    - 부서 분석업무와 역할 분담이 불명확하면 업무과다 이원화 가능성이 있다.
