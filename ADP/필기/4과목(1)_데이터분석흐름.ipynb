## 4과목 데이터 분석

### 데이터 분석 흐름
### 1. 데이터마트 구축
분석을 목적으로 `주제`$\cdot$문제정의에 관련성이 높은 데이터를 수집, 데이터 마이닝$\cdot$정형화된 패턴으로 처리, 가공해 저장하는 과정이다.
#### 1) 데이터 수집
- 데이터웨어하우스(DW)와 다른 데이터마트(DM)에서 데이터를 가져온다.
- DW에 아직 포함되지 못한 자료는 스테이징영역에서 불러와 전처리 후 기존의 데이터와 병합하거나 ODS에서 불러와 병합한다.<br>
※ ODS와 DW 차이<br>
$~~~$ ODS는 현재 또는 비교적 최신 데이터를 지속적으로 갱신해 현재 DB상태를 반영하며 DW는 현재까지의 모든 데이터를 축적 보관하고 있다.

#### 2) 데이터 마이닝
데이터 핸들링 중 하나로 여러 차원을 조합하고 새로운 변수를 생성하는 등 데이터를 가공하는 작업이다.
- 데이터 마이닝은 모델링에서도 사용되나 데이터 마트 구축 시 사전분석으로서 활용된다.
- 요약변수: 간단한 연산에 의해 데이터 정보를 요약해 담고있는 변수
  - 모델 특성에 무관하게 공통으로 사용될 수 있어 재활용성이 높다.
  - 연속형 변수를 사용할 경우 의미해석이 애매하다면 그룹화해 사용하는 것을 권장한다.
  - 예시)기간별$\cdot$채널별$\cdot$상품별 구매액$\cdot$구매횟수$\cdot$구매여부, 상품별 구매순서, 단어 빈도, 초기 행동변수, 트렌드 변수
- 파생변수: 분석가가 EDA, 사전지식을 바탕으로 특정 조건을 기준으로 한 사용자 정의 함수로 구한 변수
  - 주관적 의미가 담긴 변수로 논리적 타당성을 갖춰야 한다. 또한 분석 시점, 특정 상황에만 유의미하지 않게 대표성이 담겨야 한다.
  - 세분화, 고객행동예측, 캠페인 반응 예측에 유용하다. 
  - 예시)근무시간 구매지수, 주구매매장$\cdot$활동지역, 구매상품 다양성, 선호 가격대, 라이프 스테이지, 라이프 스타일, 행사민감, 휴면가망

#### 3) 데이터 클렌징(Data Cleansing)
정확하고 신뢰 높은 분석을 하기 위해 데이터 내 중요 변수 탐색, 오류 처리 등을 하는 작업이다.
- 변수 중요도 탐색
- 연속형 변수의 구간화
  - binning: 구간길이가 동일하게 구간화하는 방법(보통 구간수는 7개 미만) 
  - 의사결정나무: 해당 변수만 의사결정나무의 Feature로 해 얻은 분리기준값으로 구간화하는 방법
- 결측값 처리
  - 결측값은 때때로 저소득자/고소득자, 특정 거래자 등의 특수 조건을 대변하게 사용되어 그 자체가 의미있을 수도 있다.
  - 단순대치법
    - completes analysis: 결측치를 삭제하는 방법
    - 평균대치법(=비조건부평균대치법): 데이터의 평균으로 대치하는 방법
    - 회귀대치법(=조건부 평균 대치법): 회귀분석을 결과로 대치하는 방법
    - 단순확률대치법: 평균대치법의 추정량 표준오차의 과소 추정문제를 보완한 방법
      - Hot-deck 방법, nearest neighbor 방법
  - 다중대치법: m번의 대치를 통해 m개의 완전자료를 만든 뒤 분석하고 결합해 최종 완전자료를 생성하는 방법
- 이상치 처리
  - 오류에 기인한 이상치가 아닌 경우 이상치라 분류되었으나 유의미한 값일 수 있음에 제거에 보수적이어야 한다.
  - $[\mu-3\sigma, \mu+3\sigma]$(**ESD**), [geo_mean$-2.5\sigma$, geo_mean$+2.5\sigma$], [Q1$-1.5$IQR, Q3$+1.5$IQR]에 속하지 않은 데이터는 이상치로 판단한다.
  - 기하평균을 이용해 제거하거나 상하위 5%에 해당하는 데이터를 제거한다.
  - 이상치 판단의 상한과 하한 밖의 값을 각 상한, 하한으로 조정한다.
