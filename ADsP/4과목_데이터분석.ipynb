### 데이터 분석 기법
- 기본적인 R코드

- 데이터 처리<br>
  DW나 신규 시스템에 포함되지 못한 데이터는 운영시스템, 스테이징 영역, ODS에서 추출한 뒤 전처리해 DW나 DM에서 가져온 데이터와 합해 사용한다<br>
  $\bullet$ 데이터 마트(DM): 분석효과에 큰 차이를 주는 DW와 사용자의 중간층, 하나의 주제·부서 중심의 DW<br>
  - 요약변수: 분석에 맞게 종합한 변수<br>
  - 파생변수: 특정 조건·함수에 의해 값을 구성해 의미를 부여한, 작위적 변수, 논리적 타당성과 대표성을 갖춰야 함

- 데이터 가공 및 이해<br>
  $\bullet$ EDA의 4가지 주제: 저항성 강조, 잔차 계산, 자료변수의 재표현, 그래프를 통한 현시성<br>
  $\bullet$ 변수 중요도: klaR 패키지의 greedy.wilks()로 집단 내 분산/총 분산을 기준으로 변수별 종속변수로의 영향력을 평가한다.<br>
  $\bullet$ 변수 구간화: 연속형 변수는 binning, 의사결정나무에 의해 5개 정도(7개 미만)로 구간화해 모델링한다.<br>
  $\bullet$ 결측값과 이상치<br>
  $~~$ 그 자체로 의미를 갖기도 하여 부정사용방지, 침입탐지, 의료 등에 사용하기도 한다.<br>
  $~~$ 반드시 Bad Data인지 고려 후 처리해야 하며 20% 이상인 경우 해당 변수를 제외시킨다.<br>
   - 결측치 처리<br>
   $\bullet$ 단순 대치법(CA(제거), 비조건부·조건부(회귀분석)평균대치법, 단순확률대치법(추정량의 표준오차 과소 추정문제 보완))<br>
   $\bullet$ 다중 대치법: 여러 번의 단순 대치를 거쳐 가중적인 방법으로 대치하는 방법, 대치→분석→결합<br>
   $\bullet$ 이상치 처리: 데이터 손실율과 이상치 의미를 고려해 분석에서는 제외하고 추후 별도의 분석을 한다.<br>
   $~~$ ▹인식: ESD(𝜇±3𝜎 외의 값), 기하평균(기하평균±2.5𝜎 외의 값), Boxplot 외의 값<br>
   $~~$ ▹처리: 상·하한값으로 대체, 기하평균을 통한 제외, 상하위 n% 데이터 제외

### 통계분석
- 표본추출방법<br>
  $\bullet$ 단순랜덤 추출법: 동일한 선택확률로 임의의 n개를 추출하는 방법<br>
  $\bullet$ 계통추출법: n개의 구간으로 나누어 n개의 각 구간 속 k개 데이터 중 하나를 택해 n개를 추출하는 방법<br>
  ※ 표본편의: 표본추출방법에 의해 모집단을 대표하지 않는 대상이 우선적으로 추출된 오차로 확률화로 해결할 수 있다.<br>
  $~~~$ 표본오차: 모집단을 대표할 수 있는 표본단위들이 추출되지 못해 추정 시 발생하는 오류
- 확률분포<br>
  $\bullet$ 이항분포(n≥30, 𝚸≈0.5), t분포(n≥30), F분포(DF→∞)는 중심극한정리에 의해 정규분포에 근사한다.<br>
  $\bullet$ t분포: 두 집단 간 평균 동일성 검정과 모분산이 알려지지 않았을 때 모평균 가설검정, 상관계수 가설검정(H1: 𝛾≠0),<br>
  $~~\qquad$ 회귀분석의 회귀계수 유의성검정의 검정통계량<br>
  $\bullet$ 카이제곱분포: 두 집단 간의 동질성 검정, 모평균과 모분산이 알려지지 않았을 때 모분산 가설검정, <br>
  $~~\qquad\qquad$ 범주형 자료의 관측값과 기댓값 차이를 보는 적합성 검정의 검정통계량<br>
  $\bullet$ F분포: 두 집단 간 분산 동일성 검정, 다중선형회귀분석의 모형 유의성검정
- #### 가설검정
  $\bullet$ 유의수준(𝛼): H0가 사실일 때 기각할, H1을 지지할 오류(위험)의 확률<br>
  $\bullet$ p-value: 제1종 오류가 발생할 실제 확률, 유의수준(0.05)보다 작아야 유의미하며 신뢰구간 길이와 𝛽와 반비례한다.<br>
  $\bullet$ 검정력: 대립가설이 맞을 때 그것을 받아들이는 확률<br>
  ※ 비모수 검정: 자료수가 적거나 순서척도, 분포를 가정·제약할 수 없을 때 순위, 부호로 모집단 분포에 대해 설정한다. 
- #### 상관분석
  변수간의 선형정도, 선형방향을 분석한다. 추후 유의성 검증에 따른 p-value로 채택 혹은 기각한다.<br>
  $\bullet$ 피어슨 상관계수: 정규성이 가정되고 등간척도 이상으로 측정된 두 연속형 변수들의 상관계수<br>
  $\bullet$ 스피어만 상관계수: 서열척도인 두 변수들의 상관계수, 비모수적·비선형적인 순위상관계수
- #### 회귀분석(*lm* (종속 ~ 독립변수들))
  변수들 간의 인과관계를 밝히고 모형을 적합해 예측·추론하는 방법<br>
  $\bullet$ 가정
  - 선형성: 종속변수와 독립변수의 관계가 선형이다.
  - 다중공산성: 독립변수들 사이의 선형관계가 성립하지 않는 성질, 분산팽창요인(VIF)·상태지수로 판단한다.
  - 독립성: 독립변수와 오차의 𝛾=0이다. Durbin-Waston로 자기상관 구해 확인한다.
  - 비상관성: 오차는 확률표본이다.
  - 등분산성: 오차(𝜀)의 분산은 항상 𝜎이다. 잔차 플롯을 그렸을 때 y=0 주변으로 무작위하게 분포하면 만족한다.
  - 정상성: 오차는 정규분포를 따른다. `QQ plot, Shaprio-Wilk, Kolmogolov-Smirnov 검정`으로 정규성을 확인한다.
  
  $\bullet$ 분석 결과
  - 결정계수(=SSR/SST): 0과 1 사이의 값인 독립변수의 종속변수 변동 설명정도로 모형의 설명력을 의미한다.
  - 잔차와 종속변수의 산점도로 모형의 적합성을 회귀 진단한다. 
  
  $\bullet$ `최적회귀방정식` 선택
  - 단계적 변수선택: 전진선택법, 후진제거법, 단계선택법(전진선택법에 의해 변수를 추가하며 중요도 검토로 변수 제거)
  - 복잡도에 따른 벌점화: AIC, BIC(선택된 모형은 자료수가 늘어날 때 일치성이 참이 되는 모형이기도 함)
  

- #### 시계열 분석
  시간 그래프 그리기→추세·계절성 제거→잔차 예측→잔차 모델 적합→예측된 잔차에 추세·계절성 더해 예측<br>
  $\bullet$ 시계열 구분
  - 정상시계열: 평균, 분산, 자기공분산(시차(lag)가 변하면 다른 값으로 일정함)이 항상 일정하다.
  - 비정상시계열: 차분(평균 불균등), 변환(분산 불균등)해 정상화하여 분석 결과를 다른 시기로 일반화하게 된다.
  
  $\bullet$ 시계열 분석방법
  - 이동평균법: 계절·불규칙변동을 제거해 추세·순환변동만 가진 일정기간 데이터로 구한 이동평균으로 추세 파악·예측
  - 지수평활법: 개별 가중치(𝛼)를 부여해 불규칙변동이 제거된 전체데이터로 구한 평균으로 예측
  - 자기회귀모형(AR): 편자기상관함수(PACF)의 절단점 직전 자료까지가 현재 자료에 영향을 주는 모형
  - 이동평균모형(MA): 자기상관함수(ACF)의 절단점과 바로 전 시점의 백색잡음의 결합으로 이뤄진 정상성 시계열 모형
  - `자기회귀누적이동평균 모형(ARIMA(p,d,q))`: 비정상시계열을 d번 차분·변환해 정상화하여 AR(p)와 MA(q)를 따르는 모형
  - 분해시계열: `경향(추세)요인, 계절요인(고정주기), 순환요인(미지수 주기), 불규칙요인(잡음, 설명불가 오차)`을 분리해 분석
- 다차원 척도법<br>
  유사성/비유사성을 측정해 객체간 근접성을 소수 차원의 공간에 기하학적으로 표현하는 방법<br>
  $\bullet$ 반복을 통해 부적합도(stress)가 일정 수준(0.05)이하이게 하여 데이터를 축소한다. <br>
  $\bullet$ 계량적 MDS(구간척도나 비율척도), 비계량적 MDS(순서척도로 변환해 활용)
- 주성분 분석<br>
  전체 변동, 첫 번째 주성분과는 상관성이 없어 나머지 변동을 가장 잘 설명하는 변수들로 선형 조합한다.<br>
  $\bullet$ 상관행렬을 참조해 상관성은 적고 연관성은 높은 소수(4개 이하)의 주성분으로 차원, 데이터를 축소한다.<br>
  $\bullet$ 전체 변이 공헌도 방법(70~90%), 평균 고유값 방법(고유값 평균 이상), 누적기여율(설명력, 85% 이상) 방법, <br>
  $~~$  scree plot의 수평 유지(주성분수를 늘릴 때 얻게 되는 정보양이 급감) 바로 전 성분수를 주성분수로 선택한다.<br>
   ※ 요인분석: 목표변수와 변수 간 중요도 고려 없이 등간척도·비율척도인 변수들의 공통인자를 찾아 데이터를 축소함

### 데이터 마이닝
대용량 데이터에서 의미 있는 패턴을 파악해 예측하는 고급분석기법
- 성과분석(타당성 검증) 지표


 $\bullet$ ROC Curve: FPR(=1-특이도, F→T)와 민감도를 축으로 하여 왼쪽 상단에 위치할수록(AUROC≥0.8) 좋은 2진 분류모형척도<br>
 $\bullet$ 이익도표: 향상도 곡선에서 Lift가 급하게 빠른 속도로 감소할수록 좋은 분류모형척도
  - 반응검출율(등급 내 사건 수/전체 도수), 반응률(등급 내 사건 수/전체 내 등급 도수), Lift(향상도, 반응률/기본 향상도)
- 로지스틱 회귀분석(): 반응변수가 범주형으로 예측한 사후확률(최대우도추정량)로 분류하는 분류분석모형<br>
 $\bullet$ $\exp(\beta_1)$는 나머지 변수가 주어질 때 $x_1$이 한 단위 증가할 때 성공의 `오즈(승산비)`가 몇 배 증가하는 지를 의미한다.
- 의사결정나무: 세분화·분류·예측·차원축소 및 변수선택·교호작용효과 파악이 가능한 모형<br>
 $\bullet$ 성장단계에서 각 Node의 불순도 감소량을 가장 크게끔 반복해 깊어질수록 순수도가 높게하다 정지규칙에 따라 멈춘다.<br>
 $\bullet$ 순수도/불순도
 - `지니 지수`(CART) = $1 - \sum p^2$
 - 엔트로피 지수`(C4.5와 C5.0) = $- \sum p_i log_2 p_i$<br>
   $\bullet$ 무질서, 순수도·단일성 정도에 대한 측도이다.
 - 카이제곱 통계량(CHAID)

 $\bullet$ 비정상 잡음과 다중공산성 변수에 민감하지 않으나 과적합, 변수 중요도 파악불가, 경계선 값과 분류값 간 오차가 있다.
- 앙상블: 여러 개의 예측모형들을 조합해 하나의 최종예측모형을 만드는 방법<br>
 $\bullet$ 배깅: 모집단 분포를 몰라 데이터에서 단순랜덤 복원추출(36.8%)한 확률표본, 붓스트랩으로 예측모형을 만들어 결합<br>
 $\bullet$ 부스팅: 예측력이 약한 모형들을 가중치로 결합함에 훈련오차를 효율적으로 줄이고 예측오차가 향상된다.<br>
 $\bullet$ `랜덤포레스트`: 의사결정나무의 분산이 큼에 많은 무작위성으로 학습기들을 선형 결합한 것으로 정확도·예측력이 좋다. 
- 인공신경망 분석: 초기값과 가중치를 조정해 목적함수를 최적화하는 것을 목표로 한다.<br>
 $\bullet$ 역전파 알고리즘은 연결강도 갱신을 위해 에러의 역전파를 통해 가중치를 구하는데서 시작되어 비선형성을 극복했다.<br>
 $\bullet$ 뉴런이 입력신호의 가중치 합을 계산해 임계값과 비교하여 출력한 값이 활성화 함수에 대입돼 출력값이 나온다.
 - 활성화함수<br>
  $\bullet$ 시그모이드 함수: [0, 1]의 값이 출력되게 로짓 변환한 것으로 가중치는 로지스틱 회귀분석의 회귀계수와 유사하다.<br>
  $~~$ → 가중치를 0에 가깝게 하여 시그모이드 함수와 신경망 모형 모두 선형이길 목적한다.<br>
  $\bullet$ Softmax 함수: 표준화지수 함수라고도 하며 여러 개의 출력값(다범주의 범주별 사후확률)을 내놓는다.

 $\bullet$ 가중치를 점차 감소시키는 방법(벌점화)과 검증오차가 증가할 때 멈추는 방법(조기종료)로 과적합을 해결한다.
- 군집분석: 군집에 대한 가정 없이 거리(유사도 측도)로 군집 간 변동이 크게 여러 배타적인 집단으로 나눈다.<br>
  $\bullet$ 거리
  - 유클리드 거리: 산포 정도가 감안되지 않았으며 공통으로 점수 매긴 항목의 거리를 판단하는 측도
  - 표준화 거리: 척도·분산 차이로 인한 왜곡을 피하기 위해 각 변수의 표준편차로 척도 변환한 유클리드 거리
  - 마할라노비스 거리: 사전지식으로 구한 표본공분산으로 나눠주어 산포(상관성)를 고려해 표준화한 유클리드 거리
  - 맨하탄 거리: 최단거리를 구하는 거리이다. 
  - 자카드 거리(= 1 – 자카드 계수): 객체 교집합/객체 합집합, boolean 속성으로 두 객체 간 유사도를 측정한다.
  - 코사인 거리(= 1 – 코사인 유사도): 벡터 간 내적/벡터 크기의 곱

 $\bullet$ 계층적 군집분석(합병/분리): 시작 군집 개수 n개의 군집에서 군집 개수를 줄여나가 연속형 변수를 분석한다. 
  - 와드연결법: 군집 내 편차들의 제곱합을 고려한 방법으로 군집 간 정보의 손실을 최소화하기 위해 군집화 한다.<br>
    ※ 군집화: 최단연결법으로 관계규명한 후 그린 덴드로그램에서 가로선에 따라 적당한 군집수를 선정

 $\bullet$ `혼합 분포 군집(EM)`: 모분포가 k개의 모수적 모형의 가중합으로 표현된 모형이라 가정해 모수와 가중치를 추정한다.
