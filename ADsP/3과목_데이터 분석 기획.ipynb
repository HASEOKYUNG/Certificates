### 분석 방법론
일정 품질의 프로젝트 성공을 기업 내 정착하기 위해 프로젝트 절차, 방법, 도구, 산출물을 정리해두는 과정
- 각 단계가 기준선으로 설정돼 개별로 스탭의 결과, 품질검토 항목인 태스크들로 생성되는 계층적 프로세스이다.

▶빅데이터 분석 방법론<br>
 ①분석기획: 프로젝트 범위·정의·수행계획 수립(수행방법, 일정, 추진조직)·위험 식별 및 대응계획(회피/완화/수용)을 세움<br>
 $~~~~\bullet$ 분석유형 분류

 
 $~~~~\bullet$ 분석과제 발굴 방법론: 상향식(발산)·하향식(수렴)이 반복되는 디자인 사고(감정이입)을 최적이라 함<br>
 $~~\quad$▸하향식 접근방식: 문제가 정의된 분석을 의도한다.  <br>
 $\quad\qquad\qquad\qquad$ 예)폭포수 모델, 지도학습<br>
 $\qquad$1. 문제탐색: 가치(Why)에 중점을 두어 빠짐없이 비즈니스 문제(What)를 도출한다.<br>
 $~\qquad$▹비즈니스 모델: 업무, 제품, 고객(제공하는 채널 관점), 규제와 감사, 지원 인프라에 대한 기회 도출<br>
 $\qquad~~~$ → 거시적 관점의 메가트랜드(STEEP 변화), 경쟁자 확대(대체재(온라인)), <br>
 $~~~\qquad\quad$ 시장니즈 탐색(고객으로의 전달 채널, 이해관계자), 역량 재해석까지 고려해 탐색해야 한다.<br>
 $~\qquad$▹외부참조 모델: 가치 발굴 사례들을 모아둔 Pool에서 벤치마킹해(유즈케이스) Quick&Easy하게 도출<br>
 $\qquad$2. 문제정의: 탐색한 비지니스 문제(What)를 해결하기 위한 데이터 분석 문제(How)를 정의한다.<br>
 $\qquad$3. 해결방안탐색: 분석 문제를 수행할 기법에 대한 다양한 방안을 모색한다.<br>
 $\qquad$4. 타당성검토: 경제적 타당성(비용대비 편익), 데이터 및 기술적 타당성(가용 데이터, 역량) 등 다각적인 검토 수행 <br>

 $~\quad$▸상향식 접근방식: 데이터 자체(결합, 연관성, 유사성)에 대해 분석해 더 잘 이해하고 통찰한다. <br>
 $~~\quad$ ※ 프로토타이핑 접근법: 불완전하나 신속한 모형 제시 후 결과로 문제를 명확히 이해·분석해 필요 데이터를 구체화함
 
  $~~~~\bullet$ 분석 마스터 플랜 수립(ISP): Accuracy&Deploy(내재화), Long Term(중장기적), Problem Definition 특성을 갖는다.<br>
  $~~\quad$▸우선순위 평가: 전략적 중요도(필요성, 시급성), 비즈니스 성과 및 ROI, 실행 용이성(투자·기술 용이성)을 기준으로 함<br>
  $~~~~\quad$ - 시급성: 현재의 관점과 미래의 중장기적 관점 중 어디에 전략적 가치를 목표가치(ROI Value)로 둘 것인지 고려한다.<br>
  $~~~~\quad$ - 난이도: 투자비용(ROI 3V)과 분석 수준을 고려해 수행 어려움 정도를 의미하며 데이터 분석의 적합성 여부를 본다.<br>
  $~~\quad$▸이행계획 수립: 업무 내재화, 분석 데이터, 기술 적용 수준 등으로 분석 적용 범위 및 방식에 대해 최종 고려 후 수립<br>
  $~~~~\quad$ - 단계적 구현 로드맵(폭포수 모델): 데이터 분석체계 도입→유효성 검증(실현가능성)→확산 및 고도화(내재화 관리)<br>
  $~~~~\quad$ - 데이터 준비 단계까지는 순차적으로 진행하고 모델링 단계는 반복적으로 수행하는 세부 일정계획을 세운다.<br>
 ②데이터 준비: 분석에 필요한 데이터 정의·획득방안 수립·데이터 유형별 스토어 설계·수집·정합성 점검(품질점검)<br>
 ③데이터 분석: 분석용 데이터 준비(목적에 맞는 특정 데이터 추출·가공)·텍스트·EDA 분석을 거쳐<br>
 $\qquad\qquad~~~~$ 모델링(분리, 모델평가로 일반화 개선, 의사코드수준의 알고리즘 설명서 작성, 실적용성 검증, 품질검증)함<br>
 ④시스템 구현: 기존 시스템 분석 후 설계(알고리즘 설명서)·구현(BI 활용)·(이중)테스트(객관성, 완전성)·운영계획 수립(교육)<br>
 ⑤평가 및 전개: 모델 발전 계획·프로젝트 성과평가(정량적/정성적 성과)·종료(프로세스 지식 자산화)


### 분석 거버넌스 체계
데이터 분석을 기업의 문화로 정착하고 고도화하기 위한 과정
- 분석 수준진단: 기업의 분석수준을 객관적으로 파악해 도입을 위한 보완, 분석 유형 및 방향성 결정을 가능하게 함<br>
 $\bullet$ 분석 준비도: 분석 데이터, 분석업무 파악, 인력 및 조직, 분석기법, 분석 문화, IT 인프라으로 분석 도입수준을 파악한다.<br>
 $\bullet$ 분석 성숙도(CMMI, 능력 성숙도 통합모델)<br>
 $~~$ :도입단계(환경·시스템 구축)→활용단계→확산단계(분석 관리·공유)→최적화단계(분석 진화 후 혁신·성과 향상에 기여)

### 데이터 거버넌스
전사 차원의 모든 데이터에 대한 표준화된 관리체계 수립, 운영을 위한 프레임워크 및 저장소 구축
- 체계<br>
$\bullet$ 데이터 표준화: 데이터 표준 용어 설정(상호 검증 점검 프로세스), 명명 규칙 수립, 메타 데이터·데이터 사전 구축<br>
$\bullet$ 데이터 관리체계: 정합성·효율성을 위해 메타 데이터·데이터 사전 관리 원칙 수립, 데이터 생명 주기 관리방안<br>
$\bullet$ 데이터 저장소 관리: 메타 데이터 및 표준 데이터를 관리하기 위한 전사 차원의 저장소를 구성<br>
$\qquad\qquad\qquad\quad$ → 워크플로우 및 관리용 응용 SW 지원, 인터페이스를 통한 통제, 데이터 구조 변경에 따른 사전영향평가가 수행되어야 한다.<br>
$\bullet$ 표준화 활동: 표준 준수 여부 점검, 안정적 정착을 위한 교육, 실용성 확보를 위한 표준화 개선활동 수행

- 조직구조<br>
$\bullet$집중구조: 별도의 분석전담 조직<br>
$\bullet$기능구조: 업무 부서별 인력이 배치된 일반적인 구조, 전사적 핵심분석이 어렵고 과거 실적에 국한된 분석이 수행됨<br>
$\bullet$분산구조: 현업부서로 존재하며 업무 부서에 인력을 배치한 구조, 베스트 프랙티스 공유 가능함
