### 데이터
추론과 추정의 근거가 되며 다른 객체와의 상호관계 속에서 비로소 가치를 갖는 것
- 암묵지(내면화, 체화, 조직으로 공통화)와 형식지(표출화, 개인으로 연결화)의 상호작용에서 중요한 역할을 한다.
- `DIKW`<br>
  $\bullet$ 데이터<br>
  $\qquad$ 예) A마트는 100원, B마트는 200원에 연필을 판매한다.<br>
  $\bullet$ 정보: 데이터의 연관관계(상관관계)에서 도출된 의미<br>
  $\qquad$ 예) A마트가 B마트보다 싸다.<br>
  $\bullet$ 지식: 정보와 경험의 결합<br>
  $\qquad$ 예) A마트에서 구매해야겠다.<br>
  $\bullet$ 지혜: 지식이 축적되어 만든 창의적 아이디어<br>
  $\qquad$ 예) 다른 상품들도 A마트가 더 쌀 것이라고 판단한다.
- 데이터 유형<br>
  $\bullet$ 정량적(정형): 관계형 DB나 CSV 등의 고정된 필드에 저장됨<br>
  $\bullet$ 반정형: 메타데이터를 갖고 있으며, 연산이 불가하고 XML, HTML, JSON 등 파일로 저장, API로 제공됨<br>
  $\bullet$ 정성적(비정형): 멀티미디어(텍스트·영상·SNS), 사용자 정의 데이터로 객체 DB, 논리·물리적 DM(NoSQL)에 저장됨

### 데이터베이스
- 일반적인 특징<br>
 ▸ 변화되는 데이터(비휘발성): 현 시점의 데이터는 정확히 나타내며 삽입, 갱신, 삭제 등의 변화를 반영함, <br>
 $\qquad\qquad\qquad\qquad\qquad$ 제한을 두어 변화할 때 일관성, 유효성, 신뢰성을 보증하는 `데이터 무결성`이 있음<br>
 ▸ 통합된 데이터(데이터가 중복되어있지 않음), 저장된 데이터, 공용 데이터
- 기업내부 데이터베이스<br>
  $\bullet$ OLTP: 호스트 컴퓨터가 데이터베이스 갱신, 요구에 대해 액세스하고 즉각적으로 액세스에 맞게 처리<br>
  $\bullet$ OLAP: OLTP로 받은 데이터를 대화식으로 분석, 저장해 다차원적인 데이터를 의사결정에 활용하게 함<br>
  $\bullet$ CRM: 데이터의 분석 결과로 고객별 특성 마케팅을 해 평생고객화, 고객 가치 극대화를 목표로 함<br>
  $\bullet$ ERP: 생산성과 효율화를 위해 기업 경영자원을 하나의 통합 시스템으로 재구축함<br>
  $\bullet$ BI: 데이터 분석 결과를 기업의 의사결정에 활용하는 프로세스<br>
  $\bullet$ SCM: 수요자까지의 전달에 연관된 기업들 간의 IT로 모든 공급망 단계를 최적화해 수요자에게 기민하게 대응

※ 데이터웨어하우스: 특정 조직, 특정 업무 분야에 초점을 맞추어 의사결정을 지원하기 위해 구축된 데이터베이스<br>
※ 데이터 레이크: 가능한 모든 데이터를 정제 없이 자연스럽게, 접근하기 쉽게 저장해 둔 큰 데이터 세트

### 빅데이터
데이터 자체의 특성 변화, 3V(Volume, Velocity, Variety)에 초점을 둔 정의에서 처리, 분석 등의 기술 변화까지를, 더 나아가 인력(인재, 조직)변화를 포함하는 정의로 확장되었다.
- 플랫폼, 산업혁명의 석탄, 철 또는 21세기의 원유, 렌즈로 비유된다.
- 빅데이터가 만들어 낸 변화<br>
  $\bullet$ 사전처리→사후처리: 필요 여부를 따지지 않고 가능한 모든 데이터, 숨은 데이터까지 수집하는 방법으로 변화함<br>
  $\bullet$ 인과관계→상관관계: 발생 가능성을 확인하는 폭넓은 분석을 지향하며 이는 행동 추천에 의의가 있음<br>
  $\bullet$ 질→양: 방대한 데이터 속에서 더 많은 양질 데이터, 적은 오류 데이터를 얻음을 적용함<br>
  $\bullet$  표본조사→전수조사
- 빅데이터 활용 기본 테크닉<br>
 ▸ 유전자 알고리즘: 질문, 메커니즘을 통해 점진적으로 진화시켜 문제의 해결책을 최적화함<br>
 ▸ 연관규칙학습, 유형분석, 기계학습, 회귀분석, 감정분석, 소셜네트워크분석(사회관계망 분석)
- 위기별 통제방안<br>
  $\bullet$ 사생활 침해 $\rightarrow$ 정보제공자의 동의를 얻는 방식에서 정보사용자에게 책임을 묻는 방식으로의 전환, <br>$\qquad\qquad\quad~~$ 비식별 기술 적용(마스킹, 가명처리, 총계처리, 삭제, 범주화), 난수화<br>
  $\bullet$ 책임원칙 훼손 $\rightarrow$ 명확한 발생 결과에 대해서만 책임을 묻는 책임원칙훼손 원칙 고수, 피해 최소화 장치 마련<br>
  $\bullet$ 데이터 오용 $\rightarrow$ 부정확성, 잘못된 지표로의 사용에 인한 부당함 입증을 위해 알고리즘 접근권 허용

### 데이터 사이언티스트
- 역량<br>
  $\bullet$ Hard Skill: Analytics와 IT영역의 이론적 지식과 분석 숙련도<br>
  $\bullet$ Soft Skill: Analytics와 비즈니스분석 영역의 호기심, 창의, 비판, 직관 같은 통찰력 있는 분석과 전달, 협력<br>
  ※ 인문학 열풍을 부른 외부 환경 요소는 `컨버전스(단순세계화)→디버전스(복잡세계화), 생산→서비스(고객), 시장창조(무형자산 경쟁)`이다.
- 일차원적 분석: 급변하거나 큰 변화에 대응하지 못하고 새로운 기회를 포착하지 못해 지양해야 할 분석<br>
  $\bullet$ 금융: 프로그램트레이딩, 클레임분석, 신용점수 산정, 사기탐지, 가격책정, 고객 수익성분석<br>
  $\bullet$ 에너지: 트레이딩, 공급/수요 예측<br>
  $\bullet$ 병원: 고객 로열티, 수익 관리, 가격책정<br>
  $\bullet$ 정부: 사기 탐지, 사례 관리, 범죄 방지, 수익 최적화
