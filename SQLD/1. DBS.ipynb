### DataBase
한 조직의 여러 응용시스템들이 공유하기 위해 통합, 저장한 운영 `데이터`의 집합이다.
- DB의 조건<br>
  1)공용 데이터: 한 조직이 공동으로 소유, 유지, 사용하는 데이터<br>
  2)`통합 데이터`: 개별적인 저장장치에 위치하지 않게, 유일하게 DB에만 저장되어 있는 데이터<br>
   $\rightarrow$ 변경을 효율적으로 반영하고 모든 응용시스템이 동일한 최신 상태 데이터로 운영되게 한다.<br>
  3)저장 데이터: 실시간으로 데이터 요청을 받고 승인할 수 있도록 컴퓨터에 저장되어 있는 데이터<br>
  4)운영 데이터: 사용하고 있는, 질의가 이루어지는 데이터<br>
  ※ 사용하지 않는, 미래를 위해 데이터를 저장해두는 곳은 아카이브, DataWare House라 부른다.
- DB의 특성<br>
  1)실시간 접근성: 실시간으로 데이터 질의가 이루어짐<br>
  2)계속적인 변화: 데이터 lifecycle에 따라 계속 변화함<br>
  3)동시 사용: 여러 사람이 실시간으로 동시에 질의할 수 있음<br>
  4)내용에 의한 참조: 위치나 주소(디렉토리)가 아닌 데이터 내용으로 데이터를 호출할 수 있음
- 관계형 데이터베이스의 데이터는 table로 저장된다. table은 같은 record type(=데이터의 header, composite type)을 따르는 record instance(=행 데이터)의 집합이다. 즉, 행 내 index별 data type(=simple type)이 동일한 여러 행들의 집합이다.
- DB의 구성요소<br>
  • 개념적 구성요소(Instance): DB 설계자가 관심있는, DB 중 업무를 수행하는 데 의미있는 내용의 데이터<br>
  • 논리적 구성요소(Table): DB 사용자가 바라보는 데이터 형태<br>
  • 물리적 구성요소(File): 저장장치(하드디스크)에 위치한 데이터 형태
  - index가 존재해 데이터를 빠르게 검색할 수 있다.

### DBMS
데이터 구조와 내용 이해를 바탕으로 DB를 관리하는 `SoftWare`이다.
- 응용프로그램의 데이터 질의를 처리하던 File System은 데이터 입출력에만 관여하고 여러 응용프로그램들이 한 데이터를 공유한다는 점을 고려하지 않아 데이터 종속성(데이터 변경이 있으면 반드시 응용프로그램을 수정해야 하는 문제), 데이터 중복성(DB 내 각 데이터 내 중복된 속성이 존재하는 문제)를 일으켰다.
- 데이터 종속성 문제는 File System 대신 DBMS를 사용함으로써 데이터 속성 삭제로 기존 응용프로그램이 작동하지 않는 경우에만 응용프로그램을 수정하면 돼 많이 개선하였다.(`논리적 데이터 독립성`)
- 데이터 중복성을 해결하기 위해 무결성(DB 내 데이터간 연결관계에 의해 병합했을 때 모순되는 데이터가 없다는 성질)과 일관성(다수의 응용프로그램에서 동시에 특정 데이터를 변경하려 할 때 DB의 논리적 구조(복제 데이터)를 변경 후 활용하게 해 DB의 물리적 구조(원본)을 유지하는 성질, `물리적 데이터 독립성`)을 갖게 한다.

## DBS
응용프로그램에 DBMS를 통해 DB의 데이터를 전달하는 상호작용을 통칭한다.


#### $\blacktriangleright$ 스키마(schema, DB)
DB 구조와 제약조건을 명세한 table이다.
- 외부 스키마(subschema): 각 프로그램마다 유용한 정보만 담긴 데이터 형식
- 개념 스키마(schema): 모든 외부스키마를 통합한 데이터 형식
- 내부 스키마: 저장장치 관점에서 개념 스키마를 Byte로 표현한 데이터 형식
- 세 스키마로 구분하여 DBMS의 `데이터 독립성`, 개념스키마와 외부스키마간 t사상(mapping)에서 논리적 데이터 독립성, 개념스키마와 내부스키마간 사상에서 물리적 데이터 독립성을만족하게 되었다.<br>
  ex)개념스키마의 속성 추가로 각 외부스키마의 변경을 부르지 않음, 내부스키마의 변경(새로운 index지정, datatype변경)이 개념 스키마의 변경을 부르지 않음
  
#### $\blacktriangleright$ 데이터 언어(data language)
DBMS의 필수 기능, 데이터 정의(DDL), 데이터 조작(DML), 데이터 제어(DCL)을 모두 다루는 언어이다.
- SQL만이 유일한 데이터언어 SQL은 각 절차를 상세히 명시하는 절차적 언어보다 고도화된 수행할 명령만 선언하는 비절차적 언어이다.
- 사용자(응용프로그램)과 DBMS 사이에 데이터 검색, 갱신하는 DML을 가장 많이 사용한다.

#### $\blacktriangleright$ 사용자(user) & 데이터베이스 관리자(DBA)
사용자는 DB를 이용하려는 사람들을 아우르고 DBA는 DB를 설계하는 사람으로 DBA는 사용자의 요구 변화, 이용 추세 등을 분석해 DBMS 성능을 개선하고 성능을 감시한다.

#### $\blacktriangleright$ 데이터베이스 관리 시스템(DBMS)
스키마(record type 집합)와 저장 데이터베이스(record instance 집합, dataset)에 대한 모든 접근을 처리하는 소프트웨어이다.
