## CaseStudy

### University DB
- #### 교수 개체와 지도 개체 관계
  - 모든 교수가 지도교수를 담당하지 않기에 교수는 부분참여한다.
  - 한 명의 지도교수는 n명의 학생을 담당하기에 1:n 관계이며 담당 지도교수가 지도를 그만두면 해당 지도교수의 학생들은 지도교수를 잃기에 존재종속 관계이다. 따라서 학생은 전체참여한다.
- #### 개설과목 개체와 과목 개체 관계
  - 학기마다 개강, 폐강되는 과목이 다르기에 모든 과목을 instance로 가진 과목 개체와 별도로 개체과목 개체가 필요하다.
  - 모든 과목이 개설되는 것은 아니기에 과목은 부분참여한다.
  - 과목 개체에 있어야 개설 가능하기에 개설과목 개체는 과목 개체에 종속되는 약개체이다. 즉, 존재종속 관계에 따라 개설과목 개체는 전체참여한다.
  - 한 과목당 여러 분반으로 개설될 수 있기에 n:1로 맵핑된다.
- #### 교수 개체와 개설과목 개체 관계
  - 교수에게 배정된 과목이 폐강되는 경우 등으로 모든 교수가 강의하지 않기에 교수는 부분참여한다.
  - 교수가 강의하지 않으면 개설된 과목은 폐강되기에 개설과목은 존재종속관계에 따라 전체참여한다.
  - 한 교수당 여러 과목을 강의할 수 있기에 1:n의 관계이다.
- #### 학생 개체와 개설과목 개체
  - 학생 한 명은 여러 과목을 등록할 수 있어 개설과목도 여러 등록학생과 연결된다. 따라서 n:m 관계이다.
  - 휴학생 등은 과목을 수강하지 않기에 학생은 부분참여한다.
  - 모든 학생이 등록을 포기한 개설과목은 폐강되기에 개설과목은 존재종속 관계에 따라 전체참여한다.



### K-League DB
- #### 팀 개체와 경기장 개체
  - 팀별로 홈 경기장이 존재하기에 팀과 경기장은 1:1 관계이다. 
  - 모든 경기장이 하나의 팀과 전용 관계를 맺는 것이 아니기에 경기는 부분참여, 팀은 전체참여한다.
- #### 팀 개체와 경기 개체
  - 경기는 팀의 홈 경기장에서 한 번, 상대팀의 홈 경기장에 원정을 나가 한 번 이루어지기에 홈팀 관계, 어웨이팀 관계가 존재한다.
  - 특정 팀이 홈팀
- #### 경기장 개체와 경기 개체
  - 한 경기장에서 여러 회의 경기가 이루어지기에 1:n 관계이다.
  - 경기장이 사라지면 경기가 불가하기에 경기장과 경기는 존재종속 관계이고 경기장은 부분참여, 경기는 전체참여이다.
- #### 팀 개체와 선수 개체
  - 팀에 소속되지 않은, 계약만료 선수가 존재하기에 팀과 선수는 부분참여한다.
  - 한 팀에는 여러 선수가 소속되기에 1:n 관계이다.
  - 팀 내 선수를 구분하는 백넘버(부분키)로는 모든 선수를 구분할 수 없기에 팀 번호(FK)를 추가한다.



### Company DB
- #### 직원 개체와 직원가족 개체
  - 직원의 가족이어야만 DB에 저장할 가치가 생기기에 직원가족은 존재종속관계를 맺는다.
  - 직원의 가족이 없을 수 있기에 직원은 부분참여한다.
  - 한 직원의 직원가족수가 여러 명일 수 있기에 1:n 관계를 갖는다. 
- #### 상급자 직원 개체와 하급자 직원 개체
  - 동일한 직원 개체를 사용해 직속 상급자/하급자의 관계를 나타낸다.
  - 상급자를 맡지 않는 직원도, 하급자를 맡지 않는 직원도 존재해 두 개체 모두 부분참여한다.
  - 한 명의 상급자가 여러 하급자를 담당할 수 있어 1:n으로 맵핑된다.
- #### 직원 개체와 부서 개체
  - 직원이 소속되는 부서와 연결짓는 소속 관계, 부서를 총괄하는 직원과 연결짓는 관리 관계가 있다.
  - 한 부서에는 여러 직원이 소속되어 있기에 1:n관계이며 모든 부서는 적어도 한 명의 직원을 소속하지만 직원 중에는 부서에 소속되지 않는 대표, 인턴 등이 있다. 따라서 부서는 전체참여, 직원은 부분참여한다.
  - 총괄자가 없는 부서가 있을 수 있어 부서는 부분참여하고 총괄자가 있다면 한 명의 직원이 총괄자를 담당해 1:1 관계이다. 즉, 모든 직원이 부서의 총괄자가 아님에 직원도 부분참여한다.
- #### 직원 개체와 프로젝트 개체
  - 한 직원이 여러 프로젝트에 참여할 수 있어 프로젝트별 참여 직원도 여러 명이다. 즉, 직원과 프로젝트는 n:m 관계이다.
  - 모든 직원이 프로젝트에 참여하는 것은 아님으로 직원은 부분참여한다.
  - DB에 저장하는 프로젝트를 담당하는 직원이 없는 경우는 불가함으로 프로젝트는 전체참여한다.
- #### 부서 개체와 프로젝트 개체
  - 한 부서가 여러 프로젝트를 추진할 수 있기에 1:n 관계이다.
  - 현재 프로젝트를 담당하지 않는 부서가 있을 수 있어 부서는 부분참여한다.
  - 담당 부서가 없어지면 프로젝트도 무산되기에 프로젝트는 존재종속관계로 전체참여한다.
