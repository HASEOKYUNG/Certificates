## Conceptual Design
개체(테이블) 내 속성, 개체간 관계를 표현한 그림이다.<br>
$\bullet$ 기본 기호

$\bullet$ 매핑 정보: 한쪽 개체의 인스턴스가 다른쪽 개체의 인스턴스와 연결된 형태(대응성)
  - 일대일(1:1): 개체A의 한 인스턴스는 반드시 개체B의 한 인스턴스와만 연결됨
  - 일대다(1:n): 개체A의 한 인스턴스는 개체B의 여러 인스턴스와 연결되나 개체B의 한 인스턴스는 반드시 개체A의 한 인스턴스와만 연결됨
  - 다대다(n:m): 개체A, B의 인스턴스들은 서로의 여러 인스턴스와 연결됨
  
$\bullet$ 참여 정보
  - 부분참여: 개체 내 모든 인스턴스가 다른 개체의 인스턴스와 관계를 갖지 않는 경우
  - 전체참여: 개체 내 모든 인스턴스가 다른 개체의 인스턴스와 관계를 갖는 경우


## Logical Design
ERD를 바탕으로 변환규칙을 사용해 DB를 실제적이게 구현한다.<br>
※ 관계 스키마에 담긴 DB간 연관성을 PK-FK 관계로 해석할 줄 알아야 한다.<br>
$~~~$ 개념적 설계의 존재종속 관계는 논리적 설계의 식별관계를 의미한다.(각 설계별 다른 명칭으로 부름)

## 변환 규칙
개념적 ERD로 논리적 ERD를, 논리적 ERD로 개념적 ERD를 정보 손실없이 그리기 위한 규칙이다.
- [개념]
!)검색 효율성과 속도를 위해 테이블 수가 적게 관여하도록 설정한다.<br>
2)다대다 관계가 아닌 한 관계 테이블로 변환하지 않고 FK를 추가하도록 한다.<br>
3)다중값 속성은 개념적 설계에선 표현할 수 있으나 논리적 설계에선 설명할 수 없어 별도의 테이블로 작성한다.<br>
